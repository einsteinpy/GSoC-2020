```julia
using Roots
using Plots
```

```julia
q0 = [4., pi/3, 0.]
p0 = [0., 0.767851, 2.]
a = 0.99
end_lambda = 100.
step_size = 0.0005

fac = 1/2
mu = 1.

f(E) = -((a^4 * (E^2 - 2 * p0[1]^2) - 8 * a * E * p0[3] * q0[1] - 
    2 * q0[1] * (p0[2]^2 * (-2 + q0[1]) + p0[1]^2 * (-2 + q0[1])^2 * q0[1] - E^2 * q0[1]^3) + 
    a^2 * (2 * p0[3]^2 - 2 * p0[2]^2 - 4 * p0[1]^2 * (-2 + q0[1]) * q0[1] + 
    E^2 * q0[1] * (2 + 3 * q0[1])) + (a^2 + (-2 + q0[1]) * q0[1]) * (a^2 * E^2 * cos(2 * q0[2]) - 
    2 * p0[3]^2 * csc(q0[2])^2))/((a^2 + (-2 + q0[1]) * q0[1]) * (a^2 + 2 * q0[1]^2 + 
    a^2 * cos(2 * q0[2])))) + 2 * fac

E = find_zero(f, 1.)


params = [a, E, mu, fac]
```

```julia
result = KerrSolver.solveSystem(q0, p0, params, end_lambda, step_size)
```

```julia
plot(result[2], result[3][:, 1])
```
