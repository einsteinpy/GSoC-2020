```python
def leapfrog(ld, y, a):
    y_next = np.copy(y)
    fvec = np.zeros(8, dtype=np.float64)
    acc_temp = np.zeros(8, dtype=np.float64)

    fvec = f_vec3(ld, y_next, a)

    for i in range(4):
        y_next[i] += ld * y_next[i + 4] + 0.5 * ld * ld * fvec[i + 4]
        y_next[i + 4] += ld * fvec[i + 4]
        acc_temp[i] = fvec[i + 4]

    fvec = f_vec3(ld, y_next, a)

    for i in range(4):
        y[i] = y_next[i]
        y[i + 4] += 0.5 * ld * (acc_temp[i] + fvec[i + 4])
            
    return ld, y
```

```python
def v0(g_cov_mat, v1, v2, v3):
    g = g_cov_mat
    # Defining coefficients for quadratic equation
    A = g[0, 0]
    B = 2 * (g[0, 1] * v1 + g[0, 2] * v2 + g[0, 3] * v3)
    C = (
        (g[1, 1] * np.square(v1) + g[2, 2] * np.square(v2) + g[3, 3] * np.square(v3))
        + 2 * v1 * (g[1, 2] * v2 + g[1, 3] * v3)
        + 2 * v2 * g[2, 3] * v3
    )
    D = np.square(B) - (4 * A * C)

    v_t = (-B + np.sqrt(D)) / (2 * A)

    return v_t

def sphToCart(r, th, p):
    xs = r * np.sin(th) * np.cos(p)
    ys = r * np.sin(th) * np.sin(p)
    zs = r * np.cos(th)

    return xs, ys, zs

def u_norm(y, a):
    u = y[4:]
    r, th = y[1], y[2]

    g_dd = _g_dd(r, th, a)

    return u @ g_dd @ u

def _g_dd(r, th, a):
    r2, a2 = np.square(r), np.square(a)
    sint2, cost2 = np.square(np.sin(th)), np.square(np.cos(th))
    
    sg = r2 + a2 * cost2
    dl = r2 - 2 * r + a2

    g_dd = np.zeros((4, 4), dtype=np.float64)

    g_dd[0, 0] = -1 + 2 * r / sg
    g_dd[1, 1] = sg / dl
    g_dd[2, 2] = sg
    g_dd[3, 3] = (dl + (2 * r * (a2 + r2)) / sg) * sint2
    g_dd[0, 3] = g_dd[3, 0] = (-2 * a * r * sint2) / sg

    return g_dd


def _g_uu(r, th, a):
    g_dd = g_dd(r, th, a)
    g_uu = np.linalg.inv(g_dd)

    return g_uu
```

```python
def solveSystem3(end_lambda, step_size, pos, vel, a):
    g = _g_dd(pos[1], pos[2], a)

    v1, v2, v3 = vel

    v_t = v0(g, v1, v2, v3)
    state = np.hstack((pos, v_t, vel))
    print(state)
    
    lds = list()
    vecs = list()
    next_step = step_size

    # Taking first step
    next_step, y = leapfrog(next_step, state, a)
    print(next_step, y)

    lds.append(next_step)
    vecs.append(list(y))

    # Integrating and storing results from leapfrog(ld, y)
    while next_step <= end_lambda:
        step_size, y = leapfrog(step_size, y, a)
        next_step += step_size
        lds.append(next_step)
        vecs.append(list(y))
        print(next_step, y)

    lds = np.array(lds)
    vecs = np.array(vecs)

    return lds, vecs
```

```python
a = 0.9
rplus = 1 + np.sqrt(1 - np.square(a))
end_lambda = 200
# max_steps = 200
step_size = 0.05 # Too slow for 0.0005
pos = np.array([0, 20., np.pi / 2, np.pi / 2], dtype=np.float64)
vel = np.array([-0.2, 0., 0.002], dtype=np.float64)

arr = solveSystem3(end_lambda, step_size, pos, vel, a)
```

```python
sol = arr[1]

xc = list()
yc = list()
zc = list()

# Converting to Cartesian from Spherical Polar Coordinates
for y in sol:
    r, th, p = y[1], y[2], y[3]
    xs, ys, zs = sphToCart(r, th, p)
    xc.append(xs)
    yc.append(ys)
    zc.append(zs)
```
